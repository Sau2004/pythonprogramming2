```python
[rows,cols] = grey_img.shape;
neg_img = np.zeros((rows,cols));
for r in range(0, rows-1):
  for c in range(0, cols-1):
    neg_img[r,c] = 255-grey_img[r,c]; #Calculate negative image
#plt.imshow(neg_img, cmap=plt.cm.gray)
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax1.imshow(img2)
ax1.set_title('Original image')
ax2 = fig.add_subplot(2,2,2)
ax2.imshow(grey_img, cmap=plt.cm.gray)
ax2.set_title('Grey image')
ax3 = fig.add_subplot(2,2,3)
ax3.imshow(neg_img)
ax3.set_title('Negative image')
```

```python
c = 255/(np.log(1 + np.max(img2))) 
log_transformed = c * np.log(1 + img2)
log_transformed1 = np.array(log_transformed, dtype = np.uint8)
plt.imshow(log_transformed1, cmap=plt.cm.gray)
```
